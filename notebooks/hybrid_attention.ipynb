# get model predictions

```python
import torch
from tqdm import tqdm
from models import load_model
from config import Constants
from dataloader import get_loader
from misc.utils import to_device


@torch.no_grad()
def get_pred_caption(model, device, loader, vid):
    batch = loader.dataset.get_specific_data_by_vid_and_cap_id(vid, 0, device=device)
    outputs = model.translate_step(batch=batch, vocab=model.get_vocab(), assert_only_a_caption_per_video=True)
    pred_caption = outputs[vid][0]['caption']
    return pred_caption


@torch.no_grad()
def get_attr_attention_probs(model, device, loader, vid, text=None):
    model.eval()
    model.to(device)

    batch = loader.dataset.get_specific_data_by_vid_and_cap_id(vid, 0, text=text, device=device)
    results = model.captioner(batch)

    opt = model.get_opt()
    if 'concat' in opt.get('use_attr_type', ''):
        attr_attention_probs = results['all_inter_attentions'][0][:, :, :, -opt['use_attr_topk']:]
        assert attr_attention_probs.size(-1) == opt['use_attr_topk'], attr_attention_probs.shape
    else:
        attr_attention_probs = results['attr_attention_probs'][0]

    eos_index = batch['labels'][0, :].cpu().numpy().tolist().index(Constants.EOS)
    probs = attr_attention_probs[0, :, :eos_index+1, :].cpu().numpy() # (num_heads, seq_len, use_attr_topk)
    return probs


def run(ckpt_path):    
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    model = load_model(
        ckpt_path,
        device=device,
        strict=True,
        base_data_path=Constants.base_data_path
    )

    loader = get_loader(
        model.get_opt(), 
        mode='test',
        not_shuffle=True, 
        is_validation=True, 
    )
    
    pred_captions = {}
    attn_probs = {}

    for id in tqdm(loader.dataset.ids_set):
        vid = 'video%d'%id
        pred_caption = get_pred_caption(model, device, loader, vid)
        attr_attention_probs = get_attr_attention_probs(model, device, loader, vid, text=pred_caption)
        
        pred_captions[vid] = pred_caption
        attn_probs[vid] = attr_attention_probs.tolist()
    
    return {
        'pred_captions': pred_captions,
        'attn_probs': attn_probs,
        'bias': model.state_dict()['captioner.decoder.layers.0.inter_attention.SDPA.hybrid_bias']
    }
```

```python
path_full = f'{Constants.base_checkpoint_path}/MSRVTT/Transformer/Concept/base_ViT_VA_VAT_Nc500_Nk30_G1Lc_ViT_Nr20_bias1/best.ckpt'
path_lsg = f'{Constants.base_checkpoint_path}/MSRVTT/Transformer/Concept/base_ViT_VA_VAT_Nc500_Nk30_G0Lc_ViT_Nr20_bias1/best.ckpt'

results_full = run(path_full)
results_lsg = run(path_lsg)
```

# measure bias

```python
length = 7
n_concepts = 30
n_frames = 28
n_modality = 3

bias_video_full = results_full['bias'][:, :n_frames * n_modality].mean()
bias_concept_full = results_full['bias'][:, n_frames * n_modality:].mean()

bias_video_lsg = results_lsg['bias'][:, :n_frames * n_modality].mean()
bias_concept_lsg = results_lsg['bias'][:, n_frames * n_modality:].mean()

bias_video = [bias_video_full.cpu().numpy(), bias_video_lsg.cpu().numpy()]
bias_concept = [bias_concept_full.cpu().numpy(), bias_concept_lsg.cpu().numpy()]

print(bias_video)
print(bias_concept)
```

# measure implicit gating

```python
length = 7
n_concepts = 30
n_frames = 28
n_modality = 3

def get_gate_value(results):
    attn_probs = []
    for vid, caption  in results['pred_captions'].items():
        if len(caption.split()) == length:
            attn_probs.append(torch.FloatTensor(results['attn_probs'][vid]))
            
    attn_probs = torch.stack(attn_probs, dim=0).mean(1) # (num_samples, length + 1, use_attr_topk)
    assert attn_probs.shape[1:] == (length + 1, n_concepts)
    
    overall_attn_towards_concepts = attn_probs.sum(-1)
    overall_attn_towards_video = 1 - overall_attn_towards_concepts
    
    normalized_attn_towards_concepts = overall_attn_towards_concepts / n_concepts
    normalized_attn_towards_video = overall_attn_towards_video / (n_frames * n_modality)

    gate_value = normalized_attn_towards_concepts / (normalized_attn_towards_concepts + normalized_attn_towards_video)
    
    return gate_value.mean(0)[:length].cpu().numpy(), gate_value.std(0)[:length].cpu().numpy()
```

```python
gate_mean_full, gate_std_full = get_gate_value(results_full)
gate_mean_lsg, gate_std_lsg = get_gate_value(results_lsg)
```
